```python
del EV_Data["VIN (1-10)"]
```

```python
EV_Data["Legislative District"] = EV_Data["Legislative District"].fillna(EV_Data["Legislative District"].median())
```

```python
EV_Data["Legislative District"].isnull().sum()
```

```python
EV_Data["Postal Code"] = EV_Data["Postal Code"].fillna(0).astype(int)
EV_Data["Legislative District"] = EV_Data["Legislative District"].astype(int)
EV_Data["2020 Census Tract"] = EV_Data["2020 Census Tract"].fillna(0).astype(int)
```
